## Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pickle as pkl
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report

```

## Importing Dataset

```python
df = pd.read_csv('weather.csv')
```

## Checking and removing null values

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
```

## Checking for outliers 

```python
sns.boxplot(df['precipitation'], )
plt.show()


```

```python
sns.boxplot(df['temp_max'], )
plt.show()
```

```python
sns.boxplot(df['temp_min'])
plt.show()
```

```python
sns.boxplot(df['wind'], )
plt.show()
```

```python
for i in df['precipitation']:
    data = sorted(df['precipitation'])
    Q1 = np.percentile(data, 25) 
    Q3 = np.percentile(data, 75)
    IQR = Q3 - Q1
    low_limit = Q1 - 1.5 * IQR
    up_limit = Q3 + 1.5 * IQR  # Fixed the calculation of up_limit

    mean_value = (up_limit - low_limit) / 2
    print(mean_value)

    df['precipitation'] = np.where((df['precipitation'] < low_limit), low_limit, np.where((df['precipitation'] > up_limit), up_limit, df['precipitation']))
    df['wind'] = np.where((df['wind'] < low_limit), low_limit, np.where((df['wind'] > up_limit), up_limit, df['wind']))


```

## Spliting Independent and dependent variables

```python
X = df[[ 'precipitation','wind','temp_max', 'temp_min']]
y = df['weather']
```

```python
X
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
```

## Spliting Dataset into Testing and Training Set

```python
X_train, X_test, y_train, y_test = train_test_split(X,y_encoded, test_size=0.25, random_state=42)
```

## Checking the count in each class

```python
class_counts = y.value_counts()
class_counts.plot(kind='bar')
plt.xlabel('Class')
plt.ylabel('Number of Instances')
plt.title('Class Distribution')
plt.show()
```

```python
from collections import Counter

class_counts_y = Counter(y)
plt.figure(figsize=(8, 6))
plt.pie(class_counts_y.values(),  autopct='%1.1f%%', labels = ['fog','rain','sun','drizzle', 'snow'])
plt.title("Class Distribution after Random Oversampling")
plt.show()
```

## Balancing all classes

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE()

X_train_balanced, y_train_balanced = smote.fit_resample(X_train, y_train)
```

```python
df_resampled = pd.DataFrame({'Class': y_train_balanced})
# Plot a violin plot to show class distribution after random oversampling
plt.figure(figsize=(10, 6))
sns.violinplot(data=df_resampled, x='Class')
plt.title("Class Distribution after Random Oversampling")
plt.xlabel("Class Label")
plt.ylabel("Count")
plt.show()

```

```python

from collections import Counter

class_counts = Counter(y_train_balanced)

plt.figure(figsize=(8, 6))
plt.pie(class_counts.values(),  autopct='%1.1f%%', labels =  ['fog','rain','sun','drizzle', 'snow'])
plt.title("Class Distribution after Random Oversampling")
plt.show()

```
